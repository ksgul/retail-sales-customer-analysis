```python
# import pandas library
import pandas as pd
```

```python
# use whole path to the data base file (r"") to read "\"" in Windows
file_path = r"C:\DB_mokymai\OnlineRetail.csv"
```

```python
# try encoding until file is uploded
try:
    df = pd.read_csv(file_path, encoding='utf-8', low_memory=False)
except UnicodeDecodeError:
    df = pd.read_csv(file_path, encoding='cp1252', low_memory=False)

# after code is successful use print to announce success and withdraw characteristics of data
print("File is successfully uploaded") 
print(df.shape)
print(df.head())
print("Size:", df.shape) 
print(df.dtypes)
```

```python
# data converted to dataframe
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
# convert Quantity and UnitPrice into numbers
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], errors='coerce')
# errors automaticaly converted to "NaN"
print(df.dtypes)
```

```python
# check if there are omissions
print(df.isna().sum()) 
print()
print(df.dtypes)
```

```python
# fill missing data in Description:
df['Description'] = df['Description'].fillna("Unknown") 
```

```python
# create dataset only with customers:
df_customers = df.dropna(subset=['CustomerID']).copy()
print("All data:", df.shape)
print("Only with CustomerID:", df_customers.shape)

# add revenue column:
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df_customers['TotalPrice'] = df_customers['Quantity'] * df_customers['UnitPrice']

print(df.head())

# in this data there are returns (negative quantity)
print("Negative Quantity:", (df['Quantity'] < 0).sum())
# delete returns
df = df[df['Quantity'] > 0]
df_customers = df_customers[df_customers['Quantity'] > 0]

print("Negative Quantity:", (df['Quantity'] < 0).sum())
print("All sales without returns:", df.shape)
print("Sales with CustomerID without returns:", df_customers.shape)
```

```python
# save clean data
df.to_csv("sales_clean.csv", index=False, encoding='utf-8')
df_customers.to_csv("sales_customers_clean.csv", index=False, encoding='utf-8')

# Now from one file I have two different files:
# sales_clean.csv → for sales analysis
# sales_customers_clean.csv → for client behaviour analysis
```
